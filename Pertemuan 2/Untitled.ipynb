```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
#The universe of variable and membership functions
PTMB1 = ctrl.Antecedent(np.arange(0, 200, 0.5), 'PTMB1')
PTBA1 = ctrl.Antecedent(np.arange(0, 300, 0.5), 'PTBA1')
```

```python
PTMB1['sedikit'] = fuzz.trapmf(PTMB1.universe, [0,0,10,100])
PTMB1['sedang'] = fuzz.trimf(PTMB1.universe, [100,120,133])
PTMB1['tinggi'] = fuzz.trimf(PTMB1.universe, [133,155,200])

PTBA1['sedikit'] = fuzz.trapmf(PTBA1.universe, [0,0,10,80])
PTBA1['sedang'] = fuzz.trimf(PTBA1.universe, [80,100,120])
PTBA1['tinggi'] = fuzz.trimf(PTBA1.universe, [120,240,300])

PTMB1.view()
PTBA1.view()
```
